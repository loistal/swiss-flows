```python
import numpy as np
import pandas as pd
import re
```

```python
def simple_adapt(spark_text):
    text = spark_text.replace('}\n', '}, ')
    
    return '[' + text[:(len(text) - 2)] + ']'
```

```python
dir = 'results'
```

```python
swiss_nodes_params = [4, 6, 8, 10, 12, 14]
foreign_nodes_params = [0, 1, 2, 3]
foreign_pop_param = 15000
detect_interval_params = [2, 4, 6, 8, 10]
directed_params = [False, True]
```

```python
all_filenames = []

for i_swiss_nodes, v_swiss_nodes in enumerate(swiss_nodes_params):
    for i_foreign_nodes, v_foreign_nodes in enumerate(foreign_nodes_params):
        for i_interval, v_interval in enumerate(detect_interval_params):
            for i_directed, v_directed in enumerate(directed_params):
                flow_filename = "flows_" + str(v_swiss_nodes) + "_" + str(v_foreign_nodes) + "_" + str(foreign_pop_param) + "_" + str(v_interval) + "_" + str(v_directed) + ".json"
                node_filename = "nodes_" + str(v_swiss_nodes) + "_" + str(v_foreign_nodes) + "_" + str(foreign_pop_param) + "_" + str(v_interval) + "_" + str(v_directed) + ".json"
                
                all_filenames.append(flow_filename)
                all_filenames.append(node_filename)
                
                print(flow_filename)
                print(node_filename)
                
```

```python
for filename in all_filenames:
    file = open("results/" + filename + "/part-00000", "r")
    text = file.read()
    
    corrected = simple_adapt(text)
    
    correct_file = open("d3-files/" + filename, "w")
    correct_file.write(corrected)
    
    file.close()
    correct_file.close()
```
